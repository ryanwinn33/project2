```python
# Dependencies
import pandas as pd
```

```python
# Store filepath in a variable
file = "Data/RAW_data/Public_Assistance_Summary.xlsx"
```

```python
# Read our Data file for PAS with the pandas library
data = pd.read_excel(file)
data_df = pd.DataFrame(data)
data_df
```

```python
# print out column names
for column in data_df.columns:
    print(column)
```

```python
modifiedData = data_df[['disasterNumber', 
                        'declarationDate', 
                        'Year', 
                        'incidentType',
                        'dcc',
                        'damageCategory',
                        'county',
                        'state',
                        'totalObligated',
                        'obligatedDate']]
modifiedData
```

```python
#Dropping years
drop2010 = modifiedData[modifiedData.Year != 2010]
drop2011 = drop2010[drop2010.Year != 2011]
drop2012 = drop2011[drop2011.Year != 2012]
drop2013 = drop2012[drop2012.Year != 2013]
drop2014 = drop2013[drop2013.Year != 2014]
drop2015 = drop2014[drop2014.Year != 2015]
finalYears = drop2015[drop2015.Year != 2020]
finalYears
```

```python
#confirm years are what we wante (2016-2019)
finalYears.Year.unique()
```

```python
#find all incident types
finalYears.incidentType.unique()
```

```python
#Pull wanted columns
disasterData = finalYears[['disasterNumber',  
                        'Year', 
                        'incidentType',
                        'state']]

# rename column of declariation date to count of incidents
renamedDisaster = disasterData.rename(columns={"disasterNumber": "Count of Incidents",
                                            "incidentType": "Incident Type",
                                            "state": "State"})

#Count disasters incidents for each state (DESIRED YEARS AND ALL INCIDENTS)
countDisaster = renamedDisaster.groupby(['Year','State','Incident Type'])['Count of Incidents'].nunique()
countDisaster= pd.DataFrame(countDisaster)
countDisaster
```

```python
#Total Obligated amounts by FEMA by State (DESIRED YEARS AND ALL INCIDENTS)
obligated = finalYears[['totalObligated',  
                        'Year', 
                        'incidentType',
                        'state']]

# rename column of declariation date to count of incidents
renamedObligated = obligated.rename(columns={"totalObligated": "Total Obligated",
                                            "incidentType": "Incident Type",
                                            "state": "State"})

#Total Obligated amounts by FEMA by State (DESIRED YEARS AND ALL INCIDENTS)
totalObligated = renamedObligated.groupby(['Year','State','Incident Type']).sum()
totalObligated
```

```python
#Merge total obligated table aand disaster count table broken down by year
mergeYear = pd.concat([countDisaster, totalObligated], axis=1, sort=False)
mergeYear['Total Obligated'].round(decimals=2)
mergeYear
```

```python
mergeYear.dtypes
```

```python
#months incidents occured
date = finalYears[['declarationDate',  
                        'Year', 
                        'incidentType',
                        'state',
                        'disasterNumber']]
date
```

```python
# new data frame with split value columns 
new = date["declarationDate"].str.split("-", n = 2, expand = True) 

# making separate month column from new data frame 
date["month"]= new[1] 

#change numeric month to month name
date['month'] = pd.to_datetime(date['month'], format='%m').dt.month_name().str.slice(stop=3)

date
```

```python
# rename column of declariation date to count of incidents
renamedOccur = date.rename(columns={"disasterNumber": "Count of Incidents",
                                         "state": "State",
                                         "incidentType": "Incident Type",
                                         "month": "Month"})

# #obtain count of incidents per month for each year
monthOccur = renamedOccur.groupby(['Year','State','Incident Type','Month'])['Count of Incidents'].nunique()
monthOccur= pd.DataFrame(monthOccur)
monthOccur
```

```python
#obligated costs for each incident
cost = finalYears[['totalObligated',  
                        'Year', 
                        'incidentType',
                        'state',
                        'declarationDate']]
cost
```

```python
# new data frame with split value columns 
new = cost["declarationDate"].str.split("-", n = 2, expand = True) 

# making separate month column from new data frame 
cost["month"]= new[1] 

#change numeric month to month name
cost['month'] = pd.to_datetime(cost['month'], format='%m').dt.month_name().str.slice(stop=3)

cost
```

```python
# rename column of declariation date to count of incidents
renamedCost = cost.rename(columns={"totalObligated": "Total Obligated",
                                         "state": "State",
                                         "incidentType": "Incident Type",
                                         "month": "Month"})

#obtain count of incidents per month for each year
monthCost = renamedCost.groupby(['Year','State','Incident Type','Month']).sum()
monthCost
```

```python
#Merge total obligated table aand disaster count table broken down by year
mergeMonth = pd.concat([monthOccur, monthCost], axis=1, sort=False)
mergeMonth['Total Obligated'].round(decimals=2)
mergeMonth
```

```python
mergeMonth.dtypes
```

```python
#Export CSVs
mergeMonth.to_csv('Data/Disaster_cost_by_month.csv', float_format='%.2f')
mergeYear.to_csv('Data/Disaster_cost_by_year.csv', float_format='%.2f')
```
